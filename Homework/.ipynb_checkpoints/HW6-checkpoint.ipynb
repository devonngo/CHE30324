### 1. Calculate the moment of inertia of ${}^{14}$N=${}^{16}$O, in amu Å$^2$, the rotational energy constant, $B=\hbar^2/2I$, in kJ mol$^{-1}$, and the rotational spectral constant, $\tilde{B}=B/hc$,  in cm$^{-1}$.

Moment of Inertia $I = \mu r^2$

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt 
```

```python
N = 14.0067 # amu
O = 15.999 # amu
r = 1.15077 # bond length (angstrom)
h = 6.62607E-34 # J * s
hbar = 1.05457E-34 # J*s
Na = 6.02214E23 # avogadro's number
c = 2.99792458E8 # m/s

mu = N*O/(N+O) # amu

I = mu*r**2 # amu * A^2
print("I = {:.3f} amu * Angstrom^2\n".format(I))

I2 = I*1.660539040E-27 * 1E-20 # kg * m^2
B = hbar**2/(2*I2) # J
B_kj = (B * Na)/1000 # kJ/mol
print("The rotational energy constant = {:.3g} kJ/mol\n".format(B_kj))


B_t = (B / (h*c))/100 # cm
print('Rotational Spectral constant = {:.3f} cm^-1'.format(B_t))
```

### 2. Imagine that the NO molecule is adsorbed flat on a surface upon which it is free to rotate. Plot out the energies of the four lowest-energy rotational quantum states, in units of $\tilde{B}$, being sure to include appropriate quantum numbers and degeneracies. Also indicate the total rotational angular moment of each state, in units of $\hbar$.

```python
m = np.arange(-3, 4)
E = np.zeros(len(m))
L = np.zeros(len(m))

for i in range(len(m)):
    E[i] = hbar * c * (m[i])**2
    L[i] = m[i]

plt.plot(m, E, 'o')
plt.xlabel('Lowest Energy Quantum Rotational States')
plt.ylabel('Energy [B tilde]')
plt.grid()
plt.show()

print('In units of hbar, the rotational angular momentum is:')
for j in range(len(m)):
    print('State: {0:1g}, Momentum: {1:1g} hbar'.format(m[j], L[j]))
```

### 9. Plot the radial probability function $P_{20}(r) = r^2\lvert R_{2,0}(r)\rvert^2$. Be sure to label both axes appropriately. (*Hint:* Stick with $a_0$ as the length unit.)

```python
n = 100
P = np.zeros(n)
R = np.zeros(n)
p = np.zeros(n)

a0 = 0.529 * 1E-10
r = np.linspace(0, (20*a0), n)

maxP = 0

for i in range(len(P)):
    p[i] = r[i] / a0
    R[i] = (1/np.sqrt(2 * a0**3)) * ((1 - (p[i]/2))*np.exp(-p[i]/2))
    P[i] = (r[i]**2) * abs(R[i])**2

    # capture values
    if P[i] > maxP:
        maxP = P[i]
        maxr = r[i]

plt.plot(r, P)
plt.xlabel('Distance from Nucleus [m]')
plt.ylabel('Probability')
plt.grid()
plt.show()
```

```python
Eh = 27.212 #eV
E0 = 5.5263E-3 * np.exp(2)
r = (2 * np.exp(2)) / (np.pi * E0 * Eh) # A
rMax = r * 1E-10

print('Minimum Classical Distance = {:.3e} m'.format(rMax))
```

### 13. What is the probability of finding the electron beyond the classical distance? (Evaluate the necessary integral numerically.)

```python
from sympy import *
r = Symbol('r')

p1 = integrate(r**2 * ((1/np.sqrt(2 * a0**3)) * ((1 - ((r/a0)/2))*np.e**(-(r/a0)/2)))**2,(r, rMax, oo))
print('Probability of finding an electron beyond the classical distance is {:.4f}'.format(p1))
```

### 14. Can a 2s electron undergo an allowed transition to a lower energy electronic state? If so, what frequency (in cm$^{-1}$) of light would be emitted?

```python
# using the Rydberg eqn
n0 = 2 # init state
nf = 1 # final state
Rh = 2.18E-18 # J/H

dE = abs(-Rh * ((1/nf**2) - (1/n0**2)))
nu = dE/h # E = hv

print('The freqency emitted is {:.4e} Hz'.format(nu))
```
