### 1. Derive the Maxwell-Boltzman speed distribution for a 2-dimensional gas. (*Hint*: Think polar coordinates.)

The velocity of a two dimensional object is $v = \sqrt{v_x^2 + v_y^2}$. We know that in one dimension, the particle speed distribution is defined as $F(v_x)=(\frac{m}{2\pi kT})^\frac{1}{2}e^{\frac{-mv_x^2}{2kT}}$ 

To change to polar coordinates, we do:

$$P(v)=\int\int{P(v_x)P(v_y)v d\theta}$$

Integrating from $0$ to $2\pi$, we get:

$$P(v)=2\pi v(\frac{m}{2\pi k_BT})e^{\frac{m}{2k_BT}}=\frac{mv}{k_BT}e^{-\frac{mv^2}{2k_BT}}$$


### 2. Plot this 2-dimensional speed distribution for O$_2$ molecules at 200, 400 and 600 K.

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
# constants
kb = 1.38065E-23
m = 32/6.022E23 # molar mass / avagadro's #
T = np.array([200, 400, 600]) # Kelvin

# define speed distribution function using derived equation
def speed(v, T):
    return m*v/(kb*T)*np.exp(-m*v**2/(2*kb*T))
```

```python
# set up x-axis
v = np.linspace(0, 1500, 1500)

# set up array for temps
p = np.zeros([3, len(v)]) # set up 3 arrays within p

# iterate through
for i in range(len(v)):
    p[0][i] = speed(v[i], T[0])
    p[1][i] = speed(v[i], T[1])
    p[2][i] = speed(v[i], T[2])

# plot graphs
plt.plot(v, p[0], label="200 K")
plt.plot(v, p[1], label="400 K")
plt.plot(v, p[2], label="600 K")
plt.legend()
plt.xlabel("Speed (m/s)")
plt.ylabel("Probability")
plt.show()
```

### 3. Calculate the mean (expected value) of the speed of a 2-dimensional gas of molecules. How does your answer compare to a 3-dimensional gas?

```python
from sympy import *

m = Symbol('m',positive=True)
kb = Symbol('kb',positive=True)
T = Symbol('T',positive=True)
v = Symbol('v')

# expectation value
ev = integrate(v*(m*v/(kb*T)*exp(-m*v**2/(2*kb*T))),(v,0,oo)) 

print("The mean speed of a 2-dimensional gas is (in m/s):")
pprint(ev)

```

$$\sqrt{\frac{2\pi k_BT}{4m}}$$ which equals $$\sqrt{\frac{\pi k_BT}{2m}}$$
The mean of a 3-dimensional gas is $\frac{4}{\pi}$ times that of a 2-dimensional gas.

### 4. Calculate the mean kinetic energy of a 2-dimensional gas. How does your answer compare to a one and 3-dimensional gas?

```python
KE = integrate(1/2*v**2*m*v/(kb*T)*exp(-m*v**2/(2*kb*T)),(v,0,oo))

print("The mean KE of a 2-dimensional gas is:")
pprint(KE)
```

$$\frac{k_BT}{m}$$
The mean kinetic energy difference is based on the number of dimensions being considered multipled by $\frac{k_BT}{2m}$.
